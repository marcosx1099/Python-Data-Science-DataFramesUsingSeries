```python
import pandas as pd
import numpy as np

# #task 1 Create a Pandas Series variable to assign unique student IDs to 40 students. 
# The student IDs should start from the number 1001, increasing by 1 for each student. 
# Display the student IDs of all students.

id = np.arange(1001,1041)
ID = pd.Series(id)
ID
```

```python
#task2 The physics and mathematics scores of 40 students are provided in the form of a list in the Jupyter Notebook. 
# Store the scores in two separate Pandas Series variables and display them.

maths_marks=[35, 86, 63, 90, 50, 63, 15, 72, 15, 54, 72, 83, 58, 70, 92, 40, 59, 63, 53, 59, 55, 85, 18, 83, 67, 85, 46, 59, 77, 71, 77, 56, 79, 61, 82, 76, 60, 52, 88, 78]
physics_marks=[85, 94, 63, 59, 76, 64, 87, 55, 51, 71, 54, 94, 33, 72, 75, 67, 85, 15, 50, 50, 19, 55, 59, 65, 76, 13, 57, 70, 51, 94, 55, 50, 71, 91, 75, 51, 90, 70, 87, 56]

math = pd.Series(maths_marks,index = id)
physics= pd.Series(physics_marks,index = id)
print(f'The math students scores are listed below: \n{math}\n')

print(f'The physics students scores are listed below: \n{physics}')
```

```python
#task3 create and display the pandas series variable containing 
# the sum of the scores of physics and mathematics along with student id's

total = math + physics

print(total)

print(total.size)
```

```python
#task4 the cutoff score to qualify for round 1 is 120 marks out of the maximum marks of 200
#list the id's of the qualifying students

new_total = total[total >= 120]
print(new_total)

print(f'There are {new_total.size} students that qualified for the next round')
```

```python
#task5 Create a DataFrame containing the sum of math and physics scores of all 40 students. The column name in the DataFrame must be "round1 qualifiers".

DF_total = pd.DataFrame(total, columns = ['round1 qualifiers'])
DF_total
type(math)
```
